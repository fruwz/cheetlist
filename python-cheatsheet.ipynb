### Необходимые импорты

* `import numpy as np`
* `import pandas as pd`

Ссылка на официальную документацию [pandas](https://pandas.pydata.org/docs/getting_started/intro_tutorials/index.html).

Если вы работаете в онлайн-среде (кроме Google Colab) с ограниченными версиями библиотек, для построения графиков помимо `pandas` потребуется еще импортировать графическую библиотеку через `import matplotlib`.

### Загрузка данных

* функция `read_csv()` из `pandas`;
* функция `read_excel()` из `pandas`.

### Работа с датафреймами

Пример готового датафрейма `df`, упоминается в коде ниже:

|   | x | y  | group |
|---|:-:|----|-------|
| 0 | 1 | 10 | A     |
| 1 | 2 | 24 | B     |
| 2 | 4 | 38 | B     |
| 3 | 8 | 17 | A     |
| 4 | 5 | 15 | B     |

### Описание данных

* `df.info()`
* `df.describe()`
* `df["x"].describe()`
* `df["group"].value_counts()`

А также:

* `df["x"].min()`
* `df["x"].max()`
* `df["x"].mean()`
* `df["x"].std()`
* `df["x"].median()`
* `df["x"].quantile()`
* `df["x"].sum()`

Кроме того, есть еще базовые функции `len()` и `sum()`.

### Графики

* `df["x"].hist()`;
* `df["x"].plot.box()`;

### Сортировка, группировка, агрегирование

* `df.sort_values("x")`
* `df.sort_values(["group", "x"])`
* `df.groupby("group").describe()`
* `df.groupby("group")["x"].describe()`

### Запрос документации

* `help(len)` – пример вызова документации для базовой функции
* `help(pd.read_csv)` – пример вызова документации для функции `pandas`
* `help(df.describe)` – пример вызова документации для метода `pandas`
